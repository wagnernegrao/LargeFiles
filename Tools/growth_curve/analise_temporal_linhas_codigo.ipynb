```python
filename = '/home/wagner/Downloads/projetos_concatenados-20200221T131851Z-001/projetos_concatenados/akka.csv'
```

```python
df = pd.read_csv(filename, usecols=['Commit', 'Date', 'Project', 'blank', 'code', 'comment', 'filename', 'language'])
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
dfx = df.loc[df['filename'] == '/home/wagnernegrao/GitHub/projetos_github/akka/akka-cluster/src/main/java/akka/cluster/protobuf/msg/ClusterMessages.java']

```

```python
d = {'Date': dfx.Date.tolist(), 'Code': dfx.code.tolist()}
dfx2 = pd.DataFrame(data=d)

dfx2 = dfx2.sort_values('Date')

dfx2 = dfx2.set_index('Date')

plt.title('Evolução a cada 500 commits')
plt.ylabel('Line of Code')
plt.xlabel('Years')
plt.plot(dfx2)
plt.savefig('evolucao_500_commit.png', format='png', dpi = 300,  bbox_inches='tight')
plt.show()

```
