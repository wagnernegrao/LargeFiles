```python
def get_add(insertions):
    add = []

    for i in range(len(insertions)):
        if(i == 0):
            add.append(insertions[i])
        else:
            aux2 = insertions[i] + add[i-1]
            add.append(aux2)
    return(add)
```

```python
def get_diff(insertions, deletions):

    diff = []
    for i in range(len(insertions)):
        if(i == 0):
            diff.append(insertions[i] - deletions[i])
        else:
            aux1 = diff[i-1]
            aux2 = insertions[i] - deletions[i]
            diff.append( aux2 + aux1)
    
    return(diff)
```

```python
insertions1 = df1.Insertions.tolist()
deletions1  = df1.Deletions.tolist()


insertions2 = df2.Insertions.tolist()
deletions2  = df2.Deletions.tolist()


insertions3 = df3.Insertions.tolist()
deletions3  = df3.Deletions.tolist()


insertions4 = df4.Insertions.tolist()
deletions4  = df4.Deletions.tolist()


insertions5 = df5.Insertions.tolist()
deletions5  = df5.Deletions.tolist()

insertions6 = df6.Insertions.tolist()
deletions6  = df6.Deletions.tolist()
```

```python
add1  = get_add(insertions1)
diff1 = get_diff(insertions1, deletions1)

add2  = get_add(insertions2)
diff2 = get_diff(insertions2, deletions2)

add3  = get_add(insertions3)
diff3 = get_diff(insertions3, deletions3)

add4  = get_add(insertions4)
diff4 = get_diff(insertions4, deletions4)

add5  = get_add(insertions5)
diff5 = get_diff(insertions5, deletions5)

add6  = get_add(insertions6)
diff6 = get_diff(insertions6, deletions6)
```

```python
fig,axs = plt.subplots(3,2)

axs[0,0].plot(add1)
axs[0,0].plot(diff1)
#axs[0, 0].set_title('Unicode')

axs[0,1].plot(add2)
axs[0,1].plot(diff2)
#axs[0,1].set_title('Axis [0,0]')

axs[1,0].plot(add3)
axs[1,0].plot(diff3)
#axs[1,0].set_title('Axis [0,0]')

axs[1,1].plot(add4)
axs[1,1].plot(diff4)
#axs[1,1].set_title('Axis [0,0]')

axs[2,0].plot(add5)
axs[2,0].plot(diff5)

axs[2,1].plot(add6)
axs[2,1].plot(diff6)

for ax in axs.flat:
    ax.set(xlabel='Commits', ylabel='Insertions or Deletions')

for ax in axs.flat:
    ax.label_outer()
    
fig.set_size_inches(12.5, 10.5)


plt.savefig('plot_final.png', format='png', dpi = 300,  bbox_inches='tight')
```
